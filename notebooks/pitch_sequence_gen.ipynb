```python
from pybaseball import statcast
import pandas as pd
pd.set_option('display.max_columns', 50)
from IPython.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
```

```python
# Define the date range for the MLB 2022 season
start_date = '2022-04-07'
end_date = '2022-10-05'  # Adjust this date based on the end of the season

# Retrieve Statcast data for the specified date range
statcast_data = statcast(start_dt=start_date, end_dt=end_date)
```

```python
drop_cols = ['release_speed', 'release_pos_x','release_pos_z', 'spin_dir', 
             'spin_rate_deprecated','break_angle_deprecated', 'break_length_deprecated', 
             'zone', 'des', 'game_type','hit_location', 'bb_type', 'pfx_x',
             'pfx_z', 'plate_x', 'plate_z', 'hc_x', 'hc_y',
             'tfs_deprecated', 'tfs_zulu_deprecated',
             'vx0', 'vy0', 'vz0', 'ax', 'ay', 'az', 'sz_top', 'sz_bot',
             'hit_distance_sc', 'launch_speed', 'launch_angle', 'effective_speed',
             'release_spin_rate', 'release_extension', 'release_pos_y',
             'estimated_ba_using_speedangle', 'estimated_woba_using_speedangle',
             'woba_value', 'woba_denom', 'babip_value', 'iso_value',
             'launch_speed_angle', 'post_away_score',
             'post_home_score', 'post_bat_score', 'post_fld_score', 'spin_axis',
             'delta_home_win_exp', 'delta_run_exp']
statcast_data = statcast_data.drop(columns=drop_cols)
```

```python
second_drop_cols = ['on_3b','on_2b','on_1b','umpire','sv_id','fielder_2.1', 
                    'fielder_3', 'fielder_4', 'fielder_5', 'fielder_6', 'fielder_7', 
                    'fielder_8', 'fielder_9', 'if_fielding_alignment', 'of_fielding_alignment']
statcast_data = statcast_data.drop(columns=second_drop_cols)
```

```python
statcast_data_grouped = statcast_data.groupby(['game_pk','pitcher','at_bat_number','pitch_number'], dropna=False).first()

# Initialize an empty list to store pitch sequences
pitch_sequences = []
# Initialize an empty string to store the current pitch sequence
current_sequence = []
statcast_data_grouped['pitch_sequence'] = None

# Iterate through the DataFrame rows
for _, row in statcast_data_grouped.iterrows():
    pitch = row['pitch_type']
    event = row['events']
    #print(pitch, event)
    
    # If the event is not 'None', it indicates the end of an at-bat
    if event != None:
        # Append the current pitch sequence to the list
        current_sequence.append(pitch)
        pitch_sequences.append(current_sequence)
        #print(current_sequence)
        statcast_data_grouped.at[_,'pitch_sequence'] = current_sequence
        
        # Reset the current sequence to an empty list
        current_sequence = []
    else:
        # Add the pitch to the current sequence
        current_sequence.append(pitch)

statcast_data_grouped = statcast_data_grouped[statcast_data_grouped['events'].notna()].reset_index()
statcast_data_grouped['strikeout'] = statcast_data_grouped.apply(lambda r: 1 if r['events'] == 'strikeout' else 0, axis=1)
statcast_data_grouped
```
